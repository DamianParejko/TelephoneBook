```
"Iterator"

def licz_do(ilosc):
    numery = ['jeden', 'dwa', 'trzy', 'cztery']
    for numer in numery[:ilosc]:
        yield numer


licz_do_dwoch = lambda : licz_do(2)
licz_do_czterech = lambda : licz_do(4)

for i in licz_do_czterech():
    print(i)
```

```
"Command"

from abc import ABC, abstractmethod 
  
class Komenda(ABC): 
      
    def __init__(self, odbiorca): 
        self.odbiorca = odbiorca 
      
    def proces(self): 
        pass

class ImplementacjaKomendy(Komenda): 

    def __init__(self, odbiorca): 
        self.odbiorca = odbiorca 
  
    def proces(self): 
        self.odbiorca.wykonywanie_działan() 
  
class Odbiorca: 
      
    def wykonywanie_działan(self): 
        print('Czynność wykonywana w odbiorniku.') 
  
class Wywolanie: 

    def komenda(self, kmd): 
        self.kmd = kmd 
  
    def wykonanie(self): 
        self.kmd.proces() 

odbiorca = Odbiorca() 
kmd = ImplementacjaKomendy(odbiorca) 
wywolanie = Wywolanie() 
wywolanie.komenda(kmd) 
wywolanie.wykonanie()
```

```
"Singleton"

class Singleton(object): __instancja = None

def __new__(cls, x=None):
    if Singleton.__instancja is None:
        Singleton.__instancja = object.__new__(cls)
    Singleton.__instancja.x = x
    return Singleton.__instancja

a = Singleton() 
a.x = 'Samochod'

b = Singleton() 
b.x = 'Motor'

print(id(b)) 
print(id(a))
```

```
"Abstract factory"

import random 
  
class Kursy: 
    
    def __init__(self, kurs = None):   
        self.kurs = kurs
  
    def pokaz_kurs(self): 
  
        nazwa = self.kurs() 

        print(f'Mamy kurs {nazwa}') 
        print(f'Cena kursu równa się {nazwa.Cena()}') 
  
  
class Uczenie_Maszynowe: 
  
    def Cena(self): 
        return 11000
  
    def __str__(self): 
        return "Uczenie maszynowe"
  
class Frontend: 
  
    def Cena(self): 
        return 8000
  
    def __str__(self): 
        return "Frontend"
  
class Java: 
  
    def Cena(self): 
        return 12000
  
    def __str__(self): 
        return 'Java'
  
def random_course(): 
  
    return random.choice([Frontend, Java, Uczenie_Maszynowe])() 
  
  
wybor = Kursy(random_course)
  
for i in range(5): 
    wybor.pokaz_kurs()
```

```
"Adapter"

class Telefon():

    maks_volt = 5;

    @classmethod
    def rezultat(cls, volt):
        if volt > cls.maks_volt:
            print('Volt: {}V -- Niebezpieczeństwo!'.format(volt))
        else:
            print('Volt: {}V -- Ładowanie...'.format(volt))

    def ladowanie(self, volt):
        self.rezultat(volt)

class Europa():
    wyjsciowe_volt = 230

class Adapter():
    wejsciowe_volt = Europa.wyjsciowe_volt
    wyjsciowe_volt = Telefon.maks_volt

telefon = Telefon()
telefon.ladowanie(Adapter.wyjsciowe_volt)
```

```
"Bridge"

class API1: 
  
    def tworzenieProstopadloscianu(self, dlugosc, szerokosc, wysokosc): 
  
        print(f'dlugosc = {dlugosc}, '
              f' szerokosc = {szerokosc} i wysokosc = {wysokosc}') 
  
class API2: 
  
    def tworzenieProstopadloscianu(self, dlugosc, szerokosc, wysokosc): 
  
        print(f'dlugosc = {dlugosc}, '
              f' szerokosc = {szerokosc} i wysokosc = {wysokosc}')
  
class Prostopadloscian: 
  
    def __init__(self, dlugosc, szerokosc, wysokosc, API): 
  
        self._dlugosc = dlugosc 
        self._szerokosc = szerokosc
        self._wysokosc = wysokosc
  
        self._API = API 
  
    def produkcja(self): 
  
        self._API.tworzenieProstopadloscianu(self._dlugosc, self._szerokosc, self._wysokosc) 
  
    def rozbudowa(self, czas): 
  
        self._dlugosc = self._dlugosc * czas
        self._szerokosc = self._szerokosc * czas
        self._wysokosc = self._wysokosc * czas
  
prostopadloscian1 = Prostopadloscian(1, 2, 3, API1()) 
prostopadloscian1.produkcja() 
  
prostopadloscian2 = Prostopadloscian(19, 19, 19, API2()) 
prostopadloscian2.produkcja() 
```

```
"Decorator"

class Tekst: 
  
    def __init__(self, tekst): 
        self._tekst = tekst 
  
    def przekaz(self): 
        return self._tekst 
  
class Podkreslenie(Tekst): 
  
    def __init__(self, zmiennik): 
        self._zmiennik = zmiennik
  
    def przekaz(self): 
        return "<u>{}</u>".format(self._zmiennik.przekaz()) 
  
class Kursywa(Tekst): 
    def __init__(self, zmiennik): 
        self._zmiennik = zmiennik 
  
    def przekaz(self): 
        return "<i>{}</i>".format(self._zmiennik.przekaz()) 
  
class Pogrubienie(Tekst): 
  
    def __init__(self, zmiennik): 
        self._zmiennik = zmiennik
  
    def przekaz(self): 
        return "<b>{}</b>".format(self._zmiennik.przekaz()) 

tekst_przed = Tekst("Wzorce projektowe") 
tekst_po = Kursywa(Podkreslenie(Pogrubienie(tekst_przed))) 
  
print("przed :", tekst_przed.przekaz()) 
print("po :", tekst_po.przekaz()) 
```
